```python

#Codificado por @abouchan01
#Vamos a utilizar la creación de archivos el el programa

from tkinter import *
from tkinter import messagebox
Array= []    #Aquí vamos a ir guardando todo 

def IniciarArchivo():
    archivo = open("ag.txt","a")

def guardar():
    n = Nombre.get()
    ap = APaterno.get()
    am = AMaterno.get()
    t = Celular.get()
    Array.append(n + "$" + ap + "$" + am + "$" +t)
    escribirContacto()
    messagebox.showinfo("Saved","Contacto guardado")
    #IniciarArchivo()
   # cargar()
    mostrar()
    
    
    
def eliminar():
    eliminado = Delete.get()
    removido = False
    for elemento in Array:
        arreglo = elemento.split("$")
        if Delete.get() == arreglo[3]:
            Array.remove(elemento)
            removido = True
    escribirContacto()
    mostrar()
    if removido:
        messagebox.showinfo("eliminar","Elemento eliminado:  "+ eliminado)

    
def cargar():
    archivo = open("ag.txt","r")
    lectura = archivo.readline()
    if lectura:                        #si existe lectura
        while lectura:                       #mientras exista
            if lectura[-1]== '\n':              #si el ultimo miembro de lectura es un salto de linea
                lectura = lectura[:-1]    #lectura es igual a lo que ya tiene desde el inicio hasta su ultima posicion
                #eliminamos salto de linea.
            Array.append(lectura)
            lectura = archivo.readline()
    archivo.close()
                
        
def escribirContacto():
    archivo = open("ag.txt","w")
    Array.sort()                         ##¡¡¡¡¡¡¡¡¡¡¡¡¡¡¡¡¡AQUI VA EL CODIGO DE ORDEN!!!!!!!!!!!!!!!!!!!!!      
    for elemento in Array:
        archivo.write(elemento+ "\n")
    archivo.close()

def mostrar():                                        
    var = Text(ventana, width= 80, height = 15)
    Array.sort()                                ##¡¡¡¡¡¡¡¡¡¡¡¡¡¡¡¡¡AQUI VA EL CODIGO DE ORDEN!!!!!!!!!!!!!!!!!!!!! 
    valores = []
    var.insert(INSERT,"Apellido Paterno\t\t Apellido materno\t\tNombre\t\tCelular\n ")
    for elemento in Array:
        arreglo = elemento.split("$")
        valores.append(arreglo[3])
        var.insert(INSERT,arreglo[1]+"\t\t"+arreglo[2]+"\t\t"+arreglo[0]+"\t\t"+arreglo[3]+"\n")
    var.place(x=20,y= 250)
    spinEliminar = Spinbox(ventana,value=(valores), textvariable= Delete).place(x=570,y=70)
    if Array == []:
        spinEliminar = Spinbox(ventana,value=(valores)).place(x=570,y=70)
    var.config(state=DISABLED)
    
    
   
ventana = Tk()
ventana.geometry("800x600")
ventana.title("Agenda telefónica y Quick Sort")
colorFondo= "#c66"      #Vamos a trabajar con un color determinado de fondo
ventana.configure(background = colorFondo)
IniciarArchivo()
cargar()
APaterno = StringVar()
AMaterno = StringVar()
Nombre = StringVar()
Celular = StringVar()
Delete = StringVar()
mostrar()

    

                
                
                
lblTitle = Label(text = "¡Bienvenido a tu agenda!", font=("Helvetica",25), bg = colorFondo).place (x= 200,y=10)    #titulo
lblNombre = Label(text = "Nombre", font=("Helvetica",12), bg = colorFondo).place (x= 20,y=70)     #nombre
tbxNombre = Entry(ventana, textvariable = Nombre).place (x= 150,y=70)
lblAPaterno = Label(text = "Apellido paterno", font=("Helvetica",12), bg = colorFondo).place (x= 20,y=100)   #apellido paterno
tbxAPaterno = Entry(ventana, textvariable = APaterno).place (x= 150,y=100)
lblAMaterno = Label(text = "Apellido materno", font=("Helvetica",12), bg = colorFondo).place (x= 20,y=130)   #Apellido materno
tbxAMaterno = Entry(ventana, textvariable = AMaterno).place (x= 150,y=130)
lblCelular = Label(text = "Teléfono", font=("Helvetica",12), bg = colorFondo).place (x= 20,y=160)   #Celular
tbxCelular= Entry(ventana, textvariable = Celular).place (x= 150,y=160)
lblEliminar = Label(text = "Eliminar vía teléfono", font=("Helvetica",12), bg = colorFondo).place (x= 420,y=70)   #eliminar
spinEliminar = Spinbox(ventana, textvariable = Delete).place (x= 570,y=70)
btnGuardar = Button(ventana, text= "Guardar", command = guardar, bg= "#66c").place (x= 150, y=190)  #boton de guardar
btnDelete = Button(ventana, text= "Eliminar", command = eliminar, bg= "#66c").place (x= 420, y=100) #boton eliminar

mainloop()
```
